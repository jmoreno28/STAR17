```python

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def column(list, index) :
    col = np.empty(len(list))
    for i in range (len(list)):
        col[i] = list[i][index]
    return col

def sortbycol(list, index) :
    
    sortlist = np.argsort(column(list,index))
    nlist = np.empty([len(sortlist),len(list[0])])
    for i in range (len(list)):
        nlist[i] = list[sortlist[i]]
        
    return nlist
    
    
    
#notes: make a function for reading in text
#make a menu function that allows you to access functionns without hard coding values
#make a master sort function that will sort by different values

curve1 = np.loadtxt("/home/lwebster/research/practice/curves/curve1.txt", delimiter = ",", usecols = (2,3,4,5,6,7))
#print(column(text,4))
curve1 = sortbycol(curve1,4)
x1 = column(curve1,4)
y1 = column(curve1,0)

'''curve1 = np.loadtxt("/home/lwebster/research/practice/curves/curve1.txt", delimiter = ",", usecols = (2,6))
#print(column(text,4))
curve1 = sortbyCol(curve1,1)#np.sort(curve1,1)
x1 = column(curve1,1)
y1 = column(curve1,0)'''



plt.plot(x1,y1)





```
